Mathematische Definition der Algorithmen:

- resource_allocation_index: $\sum_{w \in \Gamma(u) \cap \Gamma(v)} \frac{1}{|\Gamma(w)|}$
- jaccard_coefficient: $\frac{|\Gamma(u) \cap \Gamma(v)|}{|\Gamma(u) \cup \Gamma(v)|}$
- adamic_adar_index: $\sum_{w \in \Gamma(u) \cap \Gamma(v)} \frac{1}{\log |\Gamma(w)|}$
- preferential_attachment: $|\Gamma(u)| |\Gamma(v)|$
- cn_soundarajan_hopcroft: $|\Gamma(u) \cap \Gamma(v)| + \sum_{w \in \Gamma(u) \cap \Gamma(v)} f(w)$
- ra_index_soundarajan_hopcroft:: $\sum_{w \in \Gamma(u) \cap \Gamma(v)} \frac{f(w)}{|\Gamma(w)|}$
- within_inter_cluster
- common_neighbor_centrality: $\alpha \cdot (|\Gamma (u){\cap }^{}\Gamma (v)|)+(1-\alpha )\cdot \frac{N}{{d}_{uv}}$
